Q4: # finding cube of a number that is divisible by 4 or 5 using for loop
num=[]
for i in range(1,101):
  num.append(i)
cube=[]
for k in num:
  c=k*k*k
  cube.append(c)
list=[]
for n in cube:
  if n%4==0  or n%5==0:
    list.append(n)
print(list)

using while loop

i=1
